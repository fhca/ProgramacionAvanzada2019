# Castillo, un juego de la primaria...

Objetivo: Dibujar el castillo, resultados, tiradas, sean válidas, decir quien ganó.

```python
"""
           P|S
         1_1|1_1
        2_22|22_2
       3_33.|.33_3
      4_4...|.444_4
     5_555..|.5555_5
"""

import numpy as np


def dibujo(pr, tamaño, jugadores, turno):
    """
    pr es una lista de dos elementos que son a su vez listas.
    tamaño = 'chico' (5 renglones) o 'grande' (10 renglones).
    Las sublistas son de tamaño "tamaño"
    jugadores = arreglo de dos nombres.
    turno = 0 o 1."""
    if tamaño == 'chico':
        tam = 5
        dim = 15 # la matrix tiene "shape" (dim, dim)
    else:
        tam = 10
        dim = 25
    "Matriz inicial de espacios en blanco."
    d = np.array([[' '] * dim] * dim, dtype=str)
    mitad = dim // 2
    "Pinta la linea divisoria entre jugadores."
    d[:, mitad] = '|'  # //= división entera
    r = 1
    d[r, mitad - 1] = jugadores[0][0]  # 1er caracter del 1er nombre
    d[r, mitad + 1] = jugadores[1][0]  # 1er caracter del 2o nombre
    r += 1
    "Pone los valores que están en pr en el castillo"
    for n in range(1, tam + 1):
        d[r, mitad - n - 2] = str(n)
        d[r, mitad - n - 1] = '_'
        d[r, mitad + n + 2] = str(n)
        d[r, mitad + n + 1] = '_'
        d[r, mitad + 1:mitad + n + 1] = '.'
        d[r, mitad - 1:mitad - n - 1:-1] = '.'
        d[r, mitad - n:mitad - n + pr[0][n - 1]] = str(n)
        d[r, mitad + n:mitad + n - pr[1][n - 1]:-1] = str(n)
        r += 1
    "Le indica al jugador si es su turno."
    if turno == 0:
        d[0, 2:7] = np.array(list('TURNO'))
    else:
        d[0, dim - 7:dim - 2] = np.array(list('TURNO'))
    return d


def imprime(d):
    """Imprime un dibujo desde una matriz d."""
    r, c = d.shape
    for ren in range(r):
        for col in range(c):
            print(d[ren, col], end=' ')
        print()


"""
prueba...
imprime(dibujo(pr=[[1, 2, 2, 1, 2], [1, 2, 2, 3, 4]], tamaño='chico',
               jugadores=['Paty', 'Samuel'], turno=1))
"""

def jugar():
    "FUNCION PRINCIPAL"
    jugadores = []
    jugadores.append(input("Nombre del primer jugador:"))
    jugadores.append(input("Nombre del segundo jugador:"))
    resp = input("""Quieres jugar con el tablero 'chico' (presiona c) o 
                         el 'grande' (presiona g):""")
    if resp == "c":
        tamaño = 'chico'
        tam = 5
    else:
        tamaño = 'grande'
        tam = 10
    
    # Muestra los datos...
    print(f"Los jugadores serán {jugadores[0]} y {jugadores[1]}. Jugarán en un tablero {tamaño}.")
    
    # Sorteando quien va primero
    r = np.random.randint(0,2) # genera 0 o 1
    jugador = [jugadores[r], jugadores[1-r]] 
    actual = 0
    print(f"El primero en jugar será {jugador[0]}")
    
    # estado inicial del castillo
    pr = np.zeros((2, tam), dtype=int)
    
    # comienza ciclo -......
    print(f"""
-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-
Juega {jugador[actual]}.
Ahora {jugador[actual]} escribe en un papel un número (del 1 al {tam}) sin que lo vea {jugador[1-actual]}...""")
    oponente = int(input(f"{jugador[actual]}, por favor, escribe que número pensó {jugador[1-actual]}:"))
    propio = int(input(f"Finalmente {jugador[actual]}, dime que número realmente escribiste, se honesto:"))
    # linea 8: verificando si propio es válido
    # verifica el rango, y verifica que el escalón no esté lleno
    if 1 <= propio <= tam and pr[actual][propio-1] < propio: 
        if oponente == propio:
            print(f"Te lo adivinaron!")
            pr[actual][propio-1] += 1
    
jugar()
```
