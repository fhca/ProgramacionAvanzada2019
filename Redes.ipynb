```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
nombres = {0:"Paty", 1:"Uriel", 2:"Samuel", 3:"Max", 4:"Abraham", 5:"David", 6:"Angel"}
# matriz de adyacencia o de incidencia

m = np.zeros((7,7))
m
```

agregando las conexiones. Poner un '1' en la posiciÃ³n (i,j),
si el renglon 'i' se conecta con la columna 'j'

```python
m[0, 3]=1  # Paty --> Max
m[1, 0]=1  # Uriel
m[1, 3]=1
m[1, 2]=1
m[2, 0]=1  # Samuel
m[2, 5]=1
m[3, 0]=1  # Max
m[3, 5]=1
m[4, 0]=1  # Abraham
m[4, 1]=1
m[4, 6]=1
m[5, 6]=1  # David
m[6, 6]=1  # Angel
m[6, 2]=1
m[6, 3]=1
m[6, 1]=1
m
```

```python
plt.matshow(m)
```

```python
copiones = nx.DiGraph(m)
nx.draw(copiones, labels=nombres)
```
