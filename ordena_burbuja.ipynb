```python
def acomoda_pares(a):
    for i in range(8):
        if a[i+1] < a[i]:
            a[i+1], a[i] = a[i], a[i+1]
    print(a)
```

```python
"""
Veamos si podemos estimar el número de vueltas poniendo el valor más grande al inicio...
"""
b = [9, 1, 1, 1, 1, 1, 1, 1, 1]  
b
```

```python
"""
En una sola "vuelta", se pone en su lugar el valor más grande. Por lo que no es un buen indicador.
"""
acomoda_pares(b)  
```

```python
# usando ciclos
c = [9, 9, 9, 9, 9, 9, 9, 9, 0]
def ordena_por_pares(a):
    n=8
    for i in range(n):
        print(f"vuelta {i+1} --> ", end="")
        acomoda_pares(a)

ordena_por_pares(c)
```

```python
# algoritmo en general para n valores

def ordena_burbuja(a):
    n = len(a)
    for vuelta in range(n-1):
        for i in range(n-vuelta-1):
            if a[i+1] < a[i]:
                a[i+1], a[i] = a[i], a[i+1]
        print(f"vuelta {vuelta+1} ---> {a}")

a = [7, 9, 2, 5, 4, 1, 3, 7, 5]
ordena_burbuja(a)
```

```python
b = [0, 9, 2, 5, 4, 1, 3, 7, 5]
ordena_burbuja(b)
```
